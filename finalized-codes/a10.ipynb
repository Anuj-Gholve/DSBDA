```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("iris.csv")
print(df.head())
```

```python
print("Feature and their types: ")
print(df.dtypes)
```

```python
plt.figure(figsize=(12,8))
x=df.drop('Id',axis=1)
for i,column in enumerate(x.columns[:-1]):
    plt.subplot(2,2,i+1)
    sns.histplot(df[column],edgecolor='black')
    plt.gca().bar_label(plt.gca().containers[0],fmt='%d')
    plt.title(f"Histogram of {column} ")
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
x=df.drop('Id',axis=1)
for i,column in enumerate(x.columns[:-1]):
    ax=plt.subplot(2,2,i+1)
    sns.boxplot(x=df[column],ax=ax,color='salmon')
    plt.title(f"Boxplot of {column}")
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

```python
for column in df.columns[:-1]:
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    IQR=q3-q1
    lower=q1-1.5*IQR
    upper=q3+1.5*IQR

    outliers= df[(df[column]<lower) | (df[column]>upper)][column]
print(f"Outlier dectected: {'Yes' if not outliers.empty else 'No'}")
print(f"Outliers values: {outliers.tolist() if not outliers.empty else 'None '}")
```
