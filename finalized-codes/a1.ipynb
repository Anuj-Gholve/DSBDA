```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

```python
df=pd.read_csv("iris.csv")
```

```python
print("First 5 rows: ")
print(df.head())
```

```python
#data preprocessing
print("Missing values: ")
print(df.isnull().sum())

print("Statistical summary: ")
print(df.describe())

print("Variables description and data types: ")
print(df.info())

print("Dimenesion: ")
print(df.shape)
```

```python
#Data Formatting and Data Normalization
print("Data types: ")
print(df.dtypes)

#change type of species to cateory
df['Species']=df['Species'].astype('category')
print("After type conversion: ")
print(df.dtypes)

#Normalization
#Scaling numeric values between 0 and 1

scaler=MinMaxScaler()
numeric_column=[
    'SepalLengthCm',
    'SepalWidthCm',
    'PetalLengthCm',
    'PetalWidthCm'
]

df[numeric_column]=scaler.fit_transform(df[numeric_column])
print("Normalized data: ")
print(df.head())
```

```python
#Turn categorical variables into quantitative variables
encoder=LabelEncoder()
df['Species']=encoder.fit_transform(df['Species'])
print("After encoding: ")
print(df.head())
print("Encoded values: ")
print(dict(zip(encoder.classes_,encoder.transform(encoder.classes_))))
```
