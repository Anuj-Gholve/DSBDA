```python
import pandas as pd
```

```python
df=pd.read_csv("iris.csv")
print("First 5 rows: ")
print(df.head())
```

```python
#summary statistics for a dataset with numeric variables grouped by one of the qualitative variable.

group_stats=df.groupby('Species').agg({
    "SepalLengthCm":["mean","median","min","max","std"],
     "SepalWidthCm":["mean","median","min","max","std"],
     "PetalLengthCm":["mean","median","min","max","std"],
     "PetalWidthCm":["mean","median","min","max","std"]
})

print("Summary statistics: ")
print(group_stats)
```

```python
#list that contains a numeric value for each category

species_group={}

for species in df['Species'].unique():
    species_group[species]=list(
        df[df['Species']==species]['SepalLengthCm']
    )

print("List: ")
for species,value in species_group.items():
    print(f"{species} : {value}")
    print()
```

```python
#statistical details of each category
species_list=df['Species'].unique()
for species in species_list:
    print("Statistical details: ")
    data=df[df['Species']==species]
    print(data.describe())
    print()
```
