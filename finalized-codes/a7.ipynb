```python
import nltk
from nltk.tokenize import *
from nltk.corpus import *
from nltk.stem import *
import re
```

```python
nltk.download("all")
```

```python
text="Hey! My name is Diya. I am a computer engineering student. I know Python,Java and C++. In my free time, I like listening music and reading books."
```

```python
var=sent_tokenize(text)
print(var)
```

```python
var=word_tokenize(text)
print(var)
```

```python
var=re.sub('[^A-Za-z]',' ',text)
print(var)
```

```python
var=set(stopwords.words('english'))
print("Stop words: ",var)
tokens=word_tokenize(text)
filter=[]
for w in tokens:
    if w not in var:
        filter.append(w)
print("Tokens: ",tokens)
print()
print("Filtered text: ",filter)
```

```python
var=["write","writing","wrote","written","read","reading"]
ps=PorterStemmer()
for w in var:
    root=ps.stem(w)
    print("Root word: ",root)
```

```python
wl=WordNetLemmatizer()
text="studies studying cries crying"
tt=word_tokenize(text)
for w in tt:
    print(f"Lemma for {w} is {wl.lemmatize(w)}")
```

```python
text="Hey! My name is Diya. I am a computer engineering student."
text=word_tokenize(text)
nltk.pos_tag(text)
```
