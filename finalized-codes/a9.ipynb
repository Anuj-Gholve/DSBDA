```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
print(df.head())
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='sex',y='age',hue='survived',data=df)
plt.title("Boxplot distribution of age with respect to sex and survival")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```
